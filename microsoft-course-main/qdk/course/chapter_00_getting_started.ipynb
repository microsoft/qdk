<h3>Reversibility: Adjoint auto-derived by the compiler</h3>

```python
%%qsharp

import Std.Diagnostics.DumpMachine;

// is Adj tells the compiler to auto-derive the inverse
operation Prepare(q : Qubit) : Unit is Adj {
    X(q);
    H(q);
    // state is now |−⟩ = (|0⟩ − |1⟩)/√2
}

operation AdjointDemo() : Unit {
    use q = Qubit();
    Prepare(q);
    Message("After Prepare — state is |−⟩:");
    DumpMachine();

    Adjoint Prepare(q);   // compiler-derived inverse: no manual implementation
    Message("After Adjoint Prepare — back to |0⟩:");
    DumpMachine();
    // q is guaranteed |0⟩ here; Reset() not needed
}

AdjointDemo()
```

<h3>State inspection: DumpMachine</h3>

```python
%%qsharp

import Std.Diagnostics.DumpMachine;
import Std.StatePreparation.PreparePureStateD;

// Prepare an arbitrary superposition and inspect the full state vector
operation StateInspection() : Unit {
    use qs = Qubit[3];
    // Equal superposition over |001⟩, |010⟩, |100⟩ (the three one-hot states)
    let coeffs = [0.0, 0.5773502691896258, 0.5773502691896258, 0.0,
                  0.5773502691896258, 0.0, 0.0, 0.0];
    PreparePureStateD(coeffs, qs);
    Message("State vector — each basis state and its amplitude:");
    DumpMachine();
    ResetAll(qs);
}

StateInspection()
```

<h3>Resource estimate for a small circuit</h3>

```python
# Resource estimate for a small circuit
import qsharp

qsharp.init(target_profile=qsharp.TargetProfile.Base)

result = qsharp.estimate("""
    operation EstimateMe() : Result[] {
        use qs = Qubit[4];
        H(qs[0]);
        for i in 1..3 { CNOT(qs[0], qs[i]); }
        MResetEachZ(qs)
    }
    EstimateMe()
""")

print(f"Physical qubits : {result['physicalCounts']['physicalQubits']:,}")
print(f"Runtime         : {result['physicalCounts']['runtime']}")
print(f"T gates         : {result['physicalCountsFormatted']['tCount']}")
print(f"Logical qubits  : {result['physicalCountsFormatted']['logicalQubits']}")
print()
print("(A 4-qubit GHZ state requires thousands of physical qubits on fault-tolerant hardware)")
```
