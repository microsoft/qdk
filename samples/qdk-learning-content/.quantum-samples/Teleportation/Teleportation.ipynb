# Quantum Teleportation

This notebook demonstrates quantum teleportation — transferring the state of one qubit
to another using entanglement and classical communication.

```python
from qdk import qsharp
```

```python
%%qsharp

import Std.Diagnostics.*;

operation Teleport(msg : Qubit, target : Qubit) : Unit {
    use here = Qubit();

    // Create entanglement between here and target.
    H(here);
    CNOT(here, target);

    // Bell measurement of msg and here.
    CNOT(msg, here);
    H(msg);

    // Corrections based on measurement.
    if M(here) == One { X(target); }
    if M(msg) == One  { Z(target); }

    Reset(here);
}
```

```python
%%qsharp

operation Main() : Result[] {
    use (msg, target) = (Qubit(), Qubit());

    // Prepare msg in |+⟩ state.
    H(msg);
    Message("Before teleportation:");
    DumpMachine();

    // Teleport.
    Teleport(msg, target);
    Message("After teleportation:");
    DumpMachine();

    let results = [M(msg), M(target)];
    ResetAll([msg, target]);
    results
}

Main()
```
