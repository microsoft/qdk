```python
from pathlib import Path

src = Path("benzene_diradical_injected_rotation_measurement_circuit_0.qasm").read_text()
```

```python
from qdk import init, TargetProfile, BitFlipNoise
from qdk.openqasm import compile

init(target_profile=TargetProfile.Base)
qir = compile(src)
```

```python
from qsharp._simulation import run_qir_gpu
from qdk.widgets import Histogram

output = run_qir_gpu(str(qir), shots=3, noise=BitFlipNoise(0.01), seed=None)
Histogram(output)
```
