```python
import os
from pathlib import Path

from qdk import init, TargetProfile, code
from qdk.qsharp import compile
from qdk.simulation import GpuSimulator
from qdk.widgets import Histogram

project_root = str(Path("./Be").resolve())
noise_dir = str(Path("./noise_tables").resolve())

print(f"Python process: {os.getpid()}")
```

```python
# Instantiate the GPU simulator
gpu_sim = GpuSimulator()
```

```python
print(f"Loading noise tables from {noise_dir}")
gpu_sim.load_noise_tables(noise_dir)
```

```python
print("Compiling QIR from project {project_root}")
init(target_profile=TargetProfile.Base, project_root=project_root)
qir = compile(code.Main.be_rcc_18Q_18L_all)
gpu_sim.set_program(qir)
```

```python
print("Running simulation...")
results = gpu_sim.run_shots(shots=100)
Histogram(results["shot_results"], labels="kets", items="top-25")
```
