## Setup

First, let's import the tools we'll use and define some Q# operations to work with throughout this unit.

```python
from qdk import qsharp
from qdk.qsharp import circuit
from qdk.widgets import Circuit

# Define a few operations we'll use throughout this unit.
OPERATIONS = """
    operation BellPair() : (Result, Result) {
        use (a, b) = (Qubit(), Qubit());
        H(a);
        CNOT(a, b);
        return (M(a), M(b));
    }

    operation GHZ(n : Int) : Result[] {
        use qs = Qubit[n];
        H(qs[0]);
        for i in 1..n-1 {
            CNOT(qs[0], qs[i]);
        }
        let results = MeasureEachZ(qs);
        ResetAll(qs);
        return results;
    }

    operation PrepareCatState(qs : Qubit[]) : Unit {
        H(qs[0]);
        for i in 1..Length(qs)-1 {
            CNOT(qs[0], qs[i]);
        }
    }
"""

qsharp.eval(OPERATIONS)

print("Operations defined.")
```

## Example: Basic circuit generation with an entry expression

`circuit()` takes an **entry expression** — a string that calls a Q# operation — and returns a `Circuit` object. Wrap it in the `Circuit` widget to render it in the notebook.


```python
# Generate a circuit from an entry expression
bell = circuit("BellPair()")
Circuit(bell)
```
